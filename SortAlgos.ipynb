# Bubble sort:
    Time : O(n^2)

```python
def bubble_sort(elements):
    for i in range(len(elements)-1):
        swapped = False
        for j in range(len(elements)-1-i):
            if elements[j]>elements[j+1]:
                elements[j],elements[j+1] = elements[j+1],elements[j]
                swapped = True
        if not swapped:
            return elements
    return elements
```

```python
l = [5,10,67,3,89,3,2,1,45,23]
```

```python
print(bubble_sort(l))
```

# Merge sort
    Dvide and conquer
    Time Cmp : O(n log n)

```python
def merge(x,y,arr):
    i=j=k=0
    while i<len(x) and j<len(y):
        if x[i]<y[j]:
            arr[k]=x[i]
            i+=1
        else:
            arr[k]=y[j]
            j+=1
        k+=1
    while i<len(x):
        arr[k]=x[i]
        i+=1
        k+=1
    while j<len(y):
        arr[k]=y[j]
        j+=1
        k+=1
                

def merge_sort(arr):
    if len(arr)<=1:
        return 
    
    left = arr[:len(arr)//2]
    right = arr[len(arr)//2:]
    
    merge_sort(left)
    merge_sort(right)
    
    return merge(left,right,arr)
```

```python
a = [3,45,2,90,1,12,48,15]
merge_sort(a)
print(a)
```

# Insertion Sort
    Time cmp: O(n^2)

```python
def insertion_sort(arr):
    for i in range(1,len(arr)):
        key = arr[i]
        j = i-1
        while j>=0 and key<arr[j]:
            arr[j],arr[j+1]=key,arr[j]
            j-=1
```

```python
arr = [4,3,2,1,12,45,0,9,34]
insertion_sort(arr)
print(arr)
```

# Quick Sort
    Divide and conquer.  Time : O(n log n)
    
    partioning schema : Hoare and Lomuto partitions
    
    Following uses Hoare Partition (Famous one)

```python
def partition(arr,start,end):
    pivot = end

    while start<end:
        while start<len(arr) and arr[start]<=arr[pivot]: #start searches for larger element thans pivot
            start+=1
        while arr[end]>arr[pivot]:            #end searches for smaller element than pivot
            end-=1
        if start<end: #swaps start and end at end 
            arr[end],arr[start]= arr[start],arr[end]
    
    arr[pivot],arr[end]=arr[end],arr[pivot]
        
    return end

def quick_sort(arr,start,end):
    if start<end:
        pivot_index = partition(arr,start,end) #gets the swapped indexed , centre of two partitions
        quick_sort(arr,0,pivot_index-1)
        quick_sort(arr,pivot_index+1,end)
    
```

```python
tests = [
        [11,9,29,7,2,15,28],
        [3, 7, 9, 11],
        [25, 22, 21, 10],
        [29, 15, 28],
        [],
        [6]
    ]

for elements in tests:
    quick_sort(elements, 0, len(elements)-1)
    print(f'sorted array: {elements}')
```

# Selection sort 
    Time : O(n^2)

```python
def selection_sort(arr):
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if arr[j]<arr[i]:
                arr[i],arr[j]=arr[j],arr[i]
```

```python
tests = [
        [11,9,29,7,2,15,28],
        [3, 7, 9, 11],
        [25, 22, 21, 10],
        [29, 15, 28],
        [],
        [6]
    ]
for test in tests:
    selection_sort(test)
    print(test)
```

# Shell sort
    Improvised version of insertion sort - Time : O(n log n)

```python
def shell_sort(arr):
    gap = len(arr)//2
    
    while gap>0:
        for i in range(gap,len(arr)):
            key = arr[i]
            j = i
            while j>=gap and key<arr[j-gap]:
                arr[j],arr[j-gap] = arr[j-gap],key
                j-=gap
        gap = gap//2
```

```python
tests = [[21,38,29,17,4,25,11,32,9],
        [11,9,29,7,2,15,28],
        [3, 7, 9, 11],
        [25, 22, 21, 10],
        [29, 15, 28],
        [],
        [6]
    ]
for test in tests:
    shell_sort(test)
    print(test)
```
